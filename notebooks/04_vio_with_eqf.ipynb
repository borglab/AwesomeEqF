# Visual-Inertial Odometry with EqF

This notebook demonstrates implementing a simplified Visual-Inertial Odometry (VIO) system using an Equivariant Filter, as proposed in [this](https://arxiv.org/abs/2205.01980) paper by van Goor, et al.

## Learning Objectives

- Understand VIO problem formulation
- Implement EqF for SE(3) pose estimation
- Fuse IMU and visual measurements
- Compare with standard EKF approach, IEKF approach

## Prerequisites

- Completed notebook 02
- Understanding of SE(3) and IMU dynamics
- Knowledge of camera projection models
